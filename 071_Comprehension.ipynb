# List Comprehension

```python
# List Comprehension 구문
# [표현식 for 항목 in 반복가능객체 (if 조건)]
```

image.png

```python
# 주어진 리스트의 원소 데이터 들을 * 3 하여 새로운 리스트 작성하기
# ex) [1, 2, 3, 4] ==> [3, 6, 9, 12]

# 일단 지금까지 배운 방법으로 한다면
a = [1, 2, 3, 4]
result = []

for n in a:
  result.append(n*3)

print(result)
```

```python
# 그러나 파이썬에는 List Comprehension 이라는 독특한 '리스트 안의 for 문'으로 간단히 리스트 작성 가능

a = [1, 2, 3, 4]

[n * 3 for n in a]
```

image.png

```python
[n for n in range(10) if n % 2 == 0]
```

```python
#  짝수에만 3을 곱하여 담고 싶다면 다음과 같이 "if 조건"을 사용할 수 있다
# [1, 2, 3, 4] --> [6, 12]

a = [1, 2, 3, 4]

[n * 3 for n in a if n % 2 == 0]
```

```python
# 아래와 같이 짝수에만 적용되게 하려면? 홀수는 그대로.
# [1, 2, 3, 4]
#  ↓  ↓  ↓  ↓
# [1, 6, 3, 12]   <- 2, 4 에만 x3 적용

a = [1, 2, 3, 4]

[
    n * 3 if n % 2 == 0 else n
    for n in a
]
```

```python
# for문 2개 이상 사용도 가능

# [표현식 for 항목1 in 반복가능객체1 if 조건1
#         for 항목2 in 반복가능객체2 if 조건2
#         ...
#         for 항목n in 반복가능객체n if 조건n]
```

```python
[
    f'{j}-{ch}'
    for j in range(2, 10, 2)
    for ch in "ABC"
]
```

```python
# 구구단의 모든 결과를 담는 리스트
"""
['2 x 1 = 2',
 '2 x 2 = 4',
 ...
 '9 x 7 = 63',
 '9 x 8 = 72',
 '9 x 9 = 81']

"""
None

[
    f'{x} x {y} = {x * y}'
    for x in range(2,10)
    for y in range(1,10)
]
```

```python
[num % 3 for num in range(10)]
```

# Set Comprehension

```python
{num % 3 for num in range(10)} # set comprehension
```

# Dict Comprehension

```python
result = {n for n in range(5)} # set
print(result, type(result))
```

```python
result = {n:n**2 for n in range(5)} # dict
print(result, type(result))
```

```python
{
    x : list(range(x))
    for x in [1, 2, 3, 4, 5]
}
```
