# groupby
  + 데이터를 특정 컬럼으로 묶어서 그룹화
  + 아래의 세 단계를 적용하여 데이터를 그룹화(groupping) (SQL의 group by 와 개념적으로는 동일, 사용법은 유사)
    - 데이터 분할 (split)
    - operation 적용 (apply)
    - 데이터 병합 (combine)
    




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dataset_dir = r'/content/drive/MyDrive/Py19.이종현/dataset'
filepath = os.path.join(dataset_dir, 'titanic.csv')
filepath
```

## groupby(),  groups 속성
 - 각 그룹과 그룹에 속한 index를 dict 형태로 표현




```python
grouped_pclass = df.groupby('Pclass')
grouped_pclass   # DataFrameGroupBy 객체 (이하 'groupby 객체'라 하겠습니다)
```

```python
# groupby 객체의 속성
#   - groups -> dict 형태
#   - key -> 컬럼의 각 값
#   -  value -> 인덱스 값들의 list
```

```python
grouped_pclass.groups   # Pclass 의 값은 1, 2, 3 세개 밖에 없었으니 그룹 개수는 3개다
```

```python
grouped_pclass.size()  # 각 group별로 담겨 있는 데이터 개수 확인
```

```python
# 성별(Sex) 로 grouping 해보기
grouped_gender = df.groupby('Sex')
grouped_gender #그룹개수 2개!
```

## index를 이용한 group by
 - index가 있는 경우, groupby 함수에 level 사용 가능

* **set_index** 함수
 - column 데이터를 index 레벨로 변경
* **reset_index** 함수
 - 인덱스 초기화

```python
df.head()
```

```python
df.set_index('Pclass')
```

```python
df.set_index(['Pclass', 'Sex'])
```

```python
df.set_index(['Pclass', 'Sex']).reset_index()
```

```python
# index 가 있는 DataFrame 을 groupby() 할때
# 'level' 이나 '함수'를 이용할수 있다.
```

```python
df.set_index('Embarked').groupby(level=0).mean()

# level = 0:  0번째 index 기준으로 grouping 발생!
```

```python
df.set_index('Age').groupby(level=0).mean()
```

```python
# 이번에는 함수를 사용해보겠습니다 .
# groupby(함수) =>
#   함수의 매개변수는 index 다
#   함수가 리턴하는 값이 grouping 기준이 된다!


# 나이대별로 생존률 구하기
```

```python
df.set_index('Age')
```

```python
# 함수 준비
import math
def age_categorize(age):
  if math.isnan(age):     # np.isnan 과 마찬가지 동작..
    return -1

  return math.floor(age / 10) * 10     # <-- 이 리턴하는 값이 곧 grouping 기준이 된다.
```

```python
# 함수 테스트
df['Age'].apply(lambda age: age_categorize(age))
```

```python
df.set_index('Age').groupby(age_categorize).mean()
```

```python
df.set_index('Age').groupby(age_categorize).mean()['Survived']
```

###  도전: 이름(Name) 의 시작하는 알파벳으로 승객수 집계

```python
# 도전) 이름(Name) 시작 알파벳별로 승객수 집계
"""
A    51
B    72
C    69
D    43
...
...
U     1
V    15
W    33
Y     7
Z     3
"""
def apha(name):
  return name[0].upper()

df.set_index('Name').groupby(apha).count()
```

## aggregate(집계) 함수 사용하기
 - groupby 결과에 '집계함수'를 적용하여 그룹별 데이터 확인 가능




```python
# aggregate 의 매개변수로 주어지는 함수는
#     입력 : n ->  리턴 :1개 값     이와같은 집계함수 형태이어야 한다

np.sum([10, 20, 30])
```

```python
df.groupby(['Pclass', 'Sex']).sum()
```

```python
df.groupby(['Pclass', 'Sex']).aggregate(np.sum)
```

```python
# 기본 집계 함수들은 문자열로 전달해주어도 동작
# 'sum', 'mean', 'count'...


```

```python
df.groupby(['Pclass', 'Sex']).aggregate('sum')
```

```python
# 여러개의 집계함수를, list 로 전달

df.groupby(['Pclass', 'Sex']).aggregate([np.mean, np.sum, np.max])

# multi-level column 결과
```

```python
df.groupby(['Pclass', 'Sex'])[['Age', 'Fare', 'Survived']].aggregate([np.mean, np.sum, np.max])
```

# 도전 : 각 객실등급별 탑승승객인원, 생존자수, 생존률은?

```python
df.groupby('Pclass')['Survived'].aggregate(['count', 'sum', 'mean']).rename(columns={'count': '승객수', 'sum': '생존자 수', 'mean': '생존률'})
```
